## III. Modeling

### Train-Val-Test

```python
# Split the data into train, validation, and test set

train = vnq['Close'].iloc[:-275]   # Train size = rest of the time
val = vnq['Close'].iloc[-275:-22]  # Val size = 1 year (253 days)
test = vnq['Close'].iloc[-22:]     # Test size = 1 month (22 days)

# train, test = model_selection.train_test_split(vnq['Close'], test_size=253)
```

```python
ut.visualize_train_val_test(train, test, val, "VNQ")
```

[pmdarima.auto_arima](https://alkaline-ml.com/pmdarima/modules/generated/pmdarima.arima.auto_arima.html)
- m = 4 quarterly
- m = 12 monthly
- m = 1 annually

```python
# Use pmdarima to find the optimal p, d, q, and P, D, Q
arima_pmd = pm.auto_arima(train,
                          d=1,
                          max_p=2,
                          max_q=2,
                          error_action='ignore', 
                          trace=True,
                          suppress_warnings=True, 
                          maxiter=100,
                          seasonal=True,
                          m=1)  # the decomposition plot shows the seasonaility is based on 1 year
```

```python
# Save the arima order
arima_order = arima_pmd.order
arima_order
```

How to build an ARIMA model using [statsmodel](https://machinelearningmastery.com/arima-for-time-series-forecasting-with-python/) library.

```python
# Change the train values into a list
history = [x for x in train] 

pred = [] # prediction
cf_left = [] # confidence interval: lower bound
cf_right = [] # confidence interval: upper bound


for t in range(len(val)):
    # Fit the model using statsmodels library
    arima = ARIMA(history, order=arima_order).fit()

    # Make prediction on val set
    forecast, stderr, conf_int = arima.forecast()
    
    # Predicted price
    prediction = forecast
    pred.append(prediction)
    
    # Confidence interval
    cf_left.append(conf_int[0][0])
    cf_right.append(conf_int[0][1])
    
    # Actual price
    actual = val[t]
    history.append(actual)
    print('predicted=%f, expected=%f' % (prediction, actual))
```

```python
ut.evaluate(val, pred)
```

```python
# Plot the prediction
ax = vnq.Close[val.index[0]:val.index[-1]].plot(figsize=(20,10), color='blue')

# Create a df for the predictions using same index as val part
df = pd.DataFrame(pred).set_index(val.index)
df.plot(ax=ax, color= 'red')

# Add the confidence interval for prediction
ax.fill_between(val.index, cf_left, cf_right, color='dimgray', alpha=0.2)

# Add the labels to the plot.
plt.legend(('Actual', 'Prediction'), loc=2)
plt.title("VNQ Val Result", fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.savefig('arima_val_result', dpi=300, bbox_inches='tight')
plt.show()
```

>**Comment:** *ARIMA Val Result
- Both MAE (0.8425) and RMSE (1.0930) are relative small
- R2 is 0.9753 which is relative high
- This indications show that the ARIMA model learn well

## IV. Evaluation

```python
# Make predictions which match with the test size (22 days)
fc, std, ci = arima.forecast(steps=22)
```

## V. Conclusion
- ARIMA model is easy and straightforward to operate. Although the validation results show great performance, the ARIMA model perform badly on holdout test set (pretend it's unseen data)
- ARIMA model may be a good fit for short-term prediction (ex. 1 day)
- For better prediction, I recommend to train the model every time when there are new data available and make short-term prediction
